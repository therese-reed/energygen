```python
import pandas as pd
df = pd.read_csv("CLEANED Aus Energy Statistics.csv", header=[0])
df
```

```python
df.columns
```

```python
melt_df = pd.melt(df, 
                  id_vars=['Area', 'Category','Fuel Type'],
                 value_vars=['1989-90', '1990-91', '1991-92',
       '1992-93', '1993-94', '1994-95', '1995-96', '1996-97', '1997-98',
       '1998-99', '1999-00', '2000-01', '2001-02', '2002-03', '2003-04',
       '2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10',
       '2010-11', '2011-12', '2012-13', '2013-14', '2014-15', '2015-16',
       '2016-17', '2017-18', '2018-19', '2019-20'], 
                  var_name='Year', 
                  value_name='Energy Generation (GWH)')
melt_df
```

```python
AES_df = melt_df.dropna(subset=['Energy Generation (GWH)'])
```
